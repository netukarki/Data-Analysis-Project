```python
#Cleaning the data
data.isnull().sum()
```

```python
exams = data.drop(['race/ethnicity', 'parental level of education'], axis=1)
```

```python
exams.head()
```

```python
#Relationship analysis
corelation = exams.corr()
```

```python
sns.heatmap(corelation, xticklabels=corelation.columns, yticklabels=corelation.columns,
            annot =True)
```

```python
sns.pairplot(exams)
```

```python
sns.relplot(x = 'math score', y ='reading score',hue ='gender', data = exams)
```

```python
sns.displot(exams['reading score'])
```

```python
sns.catplot(x='math score', kind = 'box', data = exams)
```
